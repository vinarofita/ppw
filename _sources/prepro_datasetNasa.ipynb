# Filtering Dataset web use

1. Filter Request URI mengambil HTMLnya saja
2. Filter Status mengambil nilai yg 200
3. Filter Request Method mengambil yg GET
4. Merubah Request Time menjadi tanggal dan waktunya saja
5. Filter Berdasarkan IP Address(Remote host) dan User Agent(Request Protocol)

```python
import pandas as pd # Import pandas library
df = pd.read_csv("/content/drive/MyDrive/semester 5/ppw/webuage.csv")

df.head(20)
df.info()
```

```python
# FILTER REQUEST URI (HANYA HTML)

before = len(df)
df_html1 = df[df["Request URI"].str.endswith(".html")]
after = len(df_html1)

print("=== Filter URI HTML ===")
print("Jumlah sebelum:", before)
print("Jumlah sesudah:", after)
#Simpan ke csv
df_html1.to_csv("hasil_filter_html.csv", index=False)
df_html1.head()

```

```python
# FILTER STATUS GET == 200

before = len(df)
df_status = df[df["Status"] == 200]
after = len(df_status)

print("=== Filter Status 200 ===")
print("Jumlah sebelum:", before)
print("Jumlah sesudah:", after)

#Simpan ke csv
df_status.to_csv("hasil_filter_status.csv", index=False)
df_status.head()
```

```python
# FILTER REQUEST METHOD GET

before = len(df)
df_get = df[df["Request method"] == "GET"]
after = len(df_get)

print("=== Filter Method GET ===")
print("Jumlah sebelum:", before)
print("Jumlah sesudah:", after)

df_get.to_csv("hasil_filter_get1.csv", index=False)
df_get.head()
```

```python
# Merubah Request Time menjadi tanggal dan waktunya saja

df_time = df.copy()
before = len(df_time)

df_time["Request time"] = pd.to_datetime(df_time["Request time"], format='ISO8601')
df_time["Tanggal"] = df_time["Request time"].dt.date
df_time["Waktu"] = df_time["Request time"].dt.time

after = len(df_time)

print("=== Konversi Waktu ===")
print("Jumlah sebelum:", before)
print("Jumlah sesudah:", after)
#simpan ke csv
df_time.to_csv('hasil_times.csv', index=False)
df_time.head()
```

```python
# Filter Berdasarkan IP Address(Remote host) dan User Agent(Request Protocol)

before = len(df)

df_ipua = df[
    (df["Remote host"].str.startswith("65.55")) &
    (df["Request Protocol"] == "HTTP/1.1")
]

after = len(df_ipua)

print("=== Filter IP + User Agent ===")
print("Jumlah sebelum:", before)
print("Jumlah sesudah:", after)
#simpan ke csv
df_ipua.to_csv('hasil_ipAddres.csv', index=False)
df_ipua.head()
```

# Filtering Dataset Nasa

1. Filter Request URI mengambil HTMLnya saja
2. Filter Status mengambil nilai yg 200
3. Filter Request Method mengambil yg GET
4. Merubah Request Time menjadi tanggal dan waktunya saja


```python
 from google.colab import drive
drive.mount('/content/drive')

import pandas as pd # Import pandas library
df = pd.read_csv("/content/drive/MyDrive/semester 5/ppw/data.csv")
print("Jumlah data awal :", len(df))
df.head(10)
```

## Filter Request URI mengambil HTMLnya saja

```python
# 1. Filter Request URI berisi HTML
jumlah_sebelum = len(df)

df_html = df[df["url"].str.contains(".html", case=False, na=False)]

jumlah_sesudah = len(df_html)

print("=== FILTER REQUEST URI (HTML) ===")
print("Jumlah sebelum filtering:", jumlah_sebelum)
print("Jumlah sesudah filtering:", jumlah_sesudah)

#Simpan ke csv
df_html.to_csv("hasil_filter_html.csv", index=False)

df_html.head()
```

## Filter Status mengambil nilai yg 200

```python
# 2. Filter Status 200
jumlah_sebelum = len(df)

df_response = df[df["response"] == 200]

jumlah_sesudah = len(df_response)

print("=== FILTER response 200 ===")
print("Jumlah sebelum filtering:", jumlah_sebelum)
print("Jumlah sesudah filtering:", jumlah_sesudah)

#Simpan ke csv
df_response.to_csv("hasil_filter_response.csv", index=False)

df_response.head()
```

## Filter Request Method mengambil yg GET

```python
# 3. Filter Request Method GET
jumlah_sebelum = len(df)

df_get = df[df["method"] == "GET"]

jumlah_sesudah = len(df_get)

print("=== FILTER METHOD GET ===")
print("Jumlah sebelum filtering:", jumlah_sebelum)
print("Jumlah sesudah filtering:", jumlah_sesudah)

#Simpan ke csv
df_get.to_csv("hasil_filter_get.csv", index=False)

df_get.head()
```

## Merubah Request Time menjadi tanggal dan waktunya saja

```python
# 4. Convert Request Time
jumlah_sebelum = len(df)

df_time1 = df.copy()
df_time1["time"] = pd.to_datetime(df_time1["time"], unit='s')
df_time1["time"] = df_time1["time"].dt.strftime("%Y-%m-%d %H:%M:%S")

jumlah_sesudah = len(df_time1)

print("=== KONVERSI REQUEST TIME ===")
print("Jumlah sebelum (harus sama):", jumlah_sebelum)
print("Jumlah sesudah:", jumlah_sesudah)

#Simpan ke csv
df_time1.to_csv("hasil_filter_time.csv", index=False)

df_time1.head()
```
